### This Csv cotains deaths, death rates and was obtained from data.gov

```python
import pandas as pd
import numpy as np
deaths= pd.read_csv('deaths.csv')
deaths
```

### Null Values
In this dataset, null values are represented as ".", so we must give them a null value in order to perform analysis in a way that pandas understands.

```python
deaths = deaths.replace(to_replace='.', value=np.nan)
deaths.isnull().sum()
```

## Converting data into proper type
#### Originally, the other column in this dataset that was of a numerical type was 'Year'. The column 'Deaths', which we need in order to perfrom data analysis was originally of type 'string'. Therefore, we need to turn the data in this column into a numerical type. We also need to drop the NaN values.

```python
deaths.Deaths = deaths.Deaths.dropna().astype(np.int64)
```

```python
deaths.sort_values(['Year', 'Deaths'], ascending=[True,False])
```

#### In this cell we get the years and cause of death ordered by the number of deaths for that specific disease.

```python
#deathCauses = pd.DataFrame(deaths.groupby(['Year', 'Leading_Cause']).Deaths.agg('max')).sort_values('Deaths', ascending=False)

deathCauses = pd.DataFrame(deaths.groupby(['Year', 'Leading_Cause']).Deaths.max()).reset_index()

deathCauses.sort_values('Deaths', ascending=False)
```

#### In this last Cell we get the total deaths per year. As we can see, there is very little variation in the amount of deaths each year, suggesting there is no trend in either direction (no obvious increase or decrease).

```python
pd.DataFrame(deaths.groupby('Year').Deaths.sum())
```
