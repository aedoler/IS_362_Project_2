### Dataset on violence statistics in New York. The CSV was obtained from data.gov



```python
import pandas as pd
import numpy as np
import decimal
```

**reading the dataset**

```python
violence = pd.read_csv('violence.csv')
violence.head()
```

**Using Melt**
Implemented Melt again to reshape the orignally read csv 

```python
formatted_df = pd.melt(violence,
                       ["County", "Year", "Population"],
                       var_name="Violent_Count", 
                       value_name="Total")
formatted_df = formatted_df.sort_values(by=["County"])
formatted_df.head(10)
```

**using the power of shape**

I tried using the power of stack() to get the similar long style results of Melt. In adition, I tried to store the data
to where the violent count could be looked at from the year 2015. i added the random int to do search for the data in a 365
day year.

```python
# simulate your data using stack()
columns = ['Violent_Count' + str(x) for x in np.arange(1, 1000, 1)]
sim_2015 = pd.DataFrame(np.random.randn(365, 999), index=pd.date_range('2015', periods=365, freq='D'), columns=columns)

processed_sim_2015 = sim_2015.stack().reset_index()
processed_sim_2015.columns = ['Year', 'Violent_Count', 'Total']
processed_sim_2015
```
